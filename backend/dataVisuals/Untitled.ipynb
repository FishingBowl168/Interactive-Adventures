```python
import pandas as pd
import numpy as np

# Power BI provides the 'dataset' variable automatically
df = pd.read_excel("beeswarm.xlsx")

# Define the logic exactly as requested
def assign_random_x(stage):
    if stage == 0:
        return np.random.uniform(0.7, 1.3)
    elif stage == 1:
        return np.random.uniform(1.7, 2.3)
    else:
        # Assuming stage 2 or any other value
        return np.random.uniform(2.7, 3.3)

# Replace the stage column with the jittered values
df['status'] = df['status'].apply(assign_random_x)

# Map Impact to a Size value for the bubble chart
# Scaling by 100-300 usually looks best in Power BI
df['bubble_size'] = df['impact'] * 100

# Power BI will now see the modified 'dataset
df.to_excel("Beeswarm2.xlsx", index=False)

```
